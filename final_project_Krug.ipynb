<h5><b>STEP THREE: Perform a full Descriptive statistical analysis on the dataset</b></h5>

```python
#code goes here 
import math
import statistics

#function to find standard deviation adn variance
def find_var_std_var(list):
#finding mean
    sum_list=0
    for w in list:
        sum_list = sum_list + w

    mean_list = sum_list/len(list)

#finding standard deviation and variance
    sum_var = 0
    for x in list:
        sum_var = sum_var + (x-mean_list)**2

    variance = sum_var/(len(list) - 1)
    std_dev = math.sqrt(variance)

    return variance, std_dev

#finding covariance
def find_covariance(listx, listy):
    meanx = statistics.mean(listx)
    meany = statistics.mean(listy)
    sumxy = 0
    for i in range(len(listx)):
        sumxy = sumxy + (listx[i] - meanx) * (listy[i] - meany)
    covxy = sumxy/(len(listx) - 1)
    return covxy




YearsExperience =[1.2000000000000002, 1.4000000000000001, 1.6, 2.1, 2.3000000000000003, 3.0, 3.1, 3.3000000000000003, 3.3000000000000003, 3.8000000000000003, 4.0, 4.1, 4.1, 4.199999999999999, 4.6, 5.0, 5.199999999999999, 5.3999999999999995, 6.0, 6.1, 6.8999999999999995, 7.199999999999999, 8.0, 8.299999999999999, 8.799999999999999, 9.1, 9.6, 9.7, 10.4, 10.6]
Salary = [39344.0, 46206.0, 37732.0, 43526.0, 39892.0, 56643.0, 60151.0, 54446.0, 64446.0, 57190.0, 63219.0, 55795.0, 56958.0, 57082.0, 61112.0, 67939.0, 66030.0, 83089.0, 81364.0, 93941.0, 91739.0, 98274.0, 101303.0, 113813.0, 109432.0, 105583.0, 116970.0, 112636.0, 122392.0, 121873.0]

YearsExperience_var, YearsExperience_std = find_var_std_var(YearsExperience)
Salary_var, Salary_std = find_var_std_var(Salary)
cov_YearsExperience_Salary = find_covariance(YearsExperience, Salary)

#variance, standard deviation, and mean for YearsExperience
print("The variance of Years of experience is:",round(YearsExperience_var, 3))
print("The standard deviation of Years of experience is:",round(YearsExperience_std, 3))
print("The mean of the Years of experience is:",round(statistics.mean(YearsExperience),3))
print("The median of the Years of experience is:",round(statistics.median(YearsExperience),3))
print("The minimum and maximum Years of experience are:",round(min(YearsExperience),3), "and",max(YearsExperience))
print()
#variance, standard deviation, and mean for Salary
print("The variance of Salary is:",round(Salary_var, 3))
print("The standard deviation of Salary is:",round(Salary_std, 3))
print("The mean of the Salaries is:",round(statistics.mean(Salary),3))
print("The median of the Salaries is:",round(statistics.median(Salary),3))
print("The minimum and maximum Salaries are:",min(Salary), "and", max(Salary))
print()
#covariance print statements for YearsExperience and Salary
print("The covariance of Years of experience and Salary is:",{round(cov_YearsExperience_Salary, 3)})

```

<h5><b>STEP FOUR: Perform a full Inferential statistical analysis on the dataset</b></h5>

```python
#find the mean salary
Mean_Salary = round(statistics.mean(Salary),3)
print("The mean of the Salaries is:",Mean_Salary)
#determine the standard error for the mean
std_error = Salary_std / math.sqrt(len(Salary))
print("the standard error for them mean of salaries is:",round(std_error,3))
#determine the confidence interval for the mean with a 95% confidence
z_score = 1.96 
margin_of_error = z_score * std_error
lower_bound = Mean_Salary - margin_of_error
upper_bound = Mean_Salary + margin_of_error
print("the confidence interval for mean with a 95% confidence is:", round(lower_bound,3),"for the lower bound","and", round(upper_bound,3),"for the upper bound")
#explain what the confidence interval means
print()
print("the confidence interval gives a range of values where the true average is likely to lie within. in this case if we calculate a 95% confidence interval for the mean salary, it means that there is 95% confidence that the true average salary of the entire population lies within this range")
```

<h5><b>STEP FIVE: Perform full linear regression</b></h5>

```python
#code goes here
#How much will someone with a work experience of 10years earn.
def basic_linear_regression(X, Y):
    n = len(X)
    mean_x = sum(X) / n
    mean_y = sum(Y) / n
    
    numerator = sum((X[i] - mean_x) * (Y[i] - mean_y) for i in range(n))
    denominator = sum((X[i] - mean_x) ** 2 for i in range(n))
    m = numerator / denominator
    b = mean_y - m * mean_x
    
    return m, b


slope, intercept = basic_linear_regression(YearsExperience, Salary)
years_experience = 10
predicted_salary = slope * years_experience + intercept

print("Predicted salary for 10 years of experience is: $", predicted_salary)
```
